## Section 1
### exe. 6.1.1 What are the minimum and maximum numbers of elements in a heap of height h?
<p>
$\diamond$ A complete binary tree. <br>
$maximum\,=\,{\sideset{}{_{i=0}^{h}}\sum 2^i}\,=\,{2^{h+1}-1}$<br><br>
$\diamond$ If there is only one node in the lowest level, then all the internal nodes form a complete binary tree with height (h -1).<br>
$num\_internal\,=\,2^{h}\,-\,1$<br>
$minimum\,=\,num\_internal\,+\,1\,=\,{2^h}\,-\,1\,+\,1\,=\,2^h$<br>

<hr>
</p>

### exe. 6.1.2 Show that an n-element heap has height $\lfloor\lg n\rfloor$.
<p>
Number of nodes of level i in a complete binary tree is $2^{i}\,-\,1$. Number of nodes from root to level i in a complete binary tree has is $number\,=\,{\sideset{}{_{k=0}^{i}}\sum 2^{k}}\,=\,2^{i+1}\,-\,1$. Thus, for a compelte binary tree with height h, number of nodes is $number\,=\,{\sideset{}{_{k=0}^{h}}\sum 2^{k}}\,=\,2^{h+1}\,-\,1$. That is, $h\,=\,\lg (n\,+\,1)\,-\,1$. This is the case that the most nodes a binary tree with height h can have.<br>
For another case that this n-element heap builds an almost-complete binary tree, there is only one leaf at the lowest level. Then number of nodes in this tree excepted the one in the lowest level is $number\prime\,=\,{\sideset{}{_{k=0}^{h-1}}\sum (2^{k}})\,=\,2^{h}\,-\,1$. Then the total number is $number\,=\,2^{h}\,-\,1\,+\,1\,=\,2^{h}$. That is, $h\,=\,\lg(n-1)$.<br>
Thus, $\lg (n)\,\leq\,h\,\leq\,\lg(n+1)\,-\,1\,\lt\,\lg(n+1)$. i.e., $h\,=\,\lfloor\lg n\rfloor$, since $h$ is an integer.<br>
<hr>
</p>

### exe. 6.1.3 Show that in any subtree of a max-heap, the root of the subtree contains the largest value occurring anywhere in that subtree.
<p>This follows the max-heap property.<br>
Subtree rooted by the <i>i</i>th element, according to the max-heap property, both its left-child and right-child are no greater than the element. And since this subtree is also a max-heap, values of all children nodes are smaller than or equal to their parents. And thus, the maximum of this subtree is the value of the <i>i</i>th element.<br></p>
<p><u>Following solution is from "Algs, Instructor's Manual".</u><br>
<cite>
Assume the claim is FALSE -- i.e., that there is a subtree whose root is not the largest element in the subtree. Then the maximum element is somewhere else in the subtree, possibly even at more than one location. Let $m$ be the index of the maximum (the lowest such index if the maximum appears more than once). Since the maximum is not at the root of the subtree, node $m$ has a parent. And since the parent of a node has a lower index than the node, $A[PARENT(m)]\,\lt\,A[m]$ ($m$ is the smallest of indices of the maximum). This conflits to max-heap property that $A[PARENT(m)]\,\geq\,A[m]$. Thus, the assumption is FALSE, which means that the claim is TRUE.
</cite>
<hr>
</p>

### exe. 6.1.4 Where in a max-heap might the smallest element reside, assuming that all elements are distinct?
<p>The smallest element can be any leaf, i.e., it is in subarray <big>$A[\lfloor\frac{n}{2}\,+\,1\rfloor ..n]$</big>.<br>The number of possible elements are $\lfloor\frac{n}{2}\rfloor$ (<i><u>number of internal nodes</u></i> = ${\sideset{}{_{k=0}^{h\,-\,1}}\sum 2^{k}}\,=\,2^{h}$ = <i><u>number of nodes in level h</u></i>). The running time to find this smallest element is $O(\lg n)$.<br>
<hr>
</p>

### exe. 6.1.5 Is an array that is in sorted order a min-heap?
<p>
<b>Yes. Sorted array is a min-heap.</b><br><br>
$\diamond$ The smallest element is $A[1]$, which is also the root of the heap.<br>
$\diamond$ For a certain element $A[i]$, elements $A[i*2]$ and $A[i*2+1]$ are both smaller than or equal to $A[i]$; and $node(i*2)$ and $node(i*2+1)$ are left-child and right-child of $node(i)$ in the heap, which, according to the min-heap property, should be no greater than $node(i)$.<br><br>
Thus, a sorted array is a min-heap.<br>
<hr>
</p>

### exe.6.1.6 Is the array with values [23, 17, 14, 6, 13, 10, 1, 5, 7, 12] a max-heap?
<p>Yes, this is a max-heap.<br>
Root $A[1]\,=\,23$ is the maximum element in the array.<br>
For each <i>i</i>th element, $A[i*2]$ and $A[i*2+1]$ are not greater than $A[i]$.<br>
<hr>
</p>

### exe. 6.1.7 Show that, with the array representation for storing an n-element heap, the leaves are the nodes indexed by $\lfloor\frac{n}{2}\rfloor\,+\,1$, $\lfloor\frac{n}{2}\rfloor\,+\,2$, $\cdots$, $n$.
<p>
$\diamond$ An n-element heap has at least $(2^{h\,-\,1})$ leaves and at most $(2^{h})$ leaves.<br>
$\diamond$ An n-element heap has at least $(2^{h})$ nodes and at most $(2^{h\,+\,1})$ nodes.<br>
$\diamond$ Number of internal nodes will be in range $((2^{h}\,-\,2^{h\,-\,1}\,=\,2^{h\,-\,1}),\,(2^{h\,+\,1}\,-\,2^{h}\,=\,2^{h}))$. That is, $(\lfloor\frac{n}{2}\rfloor ,\,\lfloor\frac{n\,+\,1}{2}\rfloor)$ $(number\,of\,internal\,nodes\,=\,{\sideset{}{_{k=0}^{h\,-\,1}}\sum 2^{k}}\,=\,2^{h}\,=\,number\,of\,nodes\,in\,level\,h)$.<br><br>
Thus, the indices of leaves will be $\lfloor\frac{n}{2}\rfloor\,+\,1$, $\lfloor\frac{n}{2}\rfloor\,+\,2$, $\cdots$, $n$.
</p>

### exe.6.2.2 Starting with the procedure <i>MAX-HEAPIFY</i>, write pseudocode for the procedure <i>MIN-HEAPIFY(A, i)</i>, which performs the corresponding manipulation on a min-heap. How does the running time of <i>MIN-HEAPIFY</i> compare to that of <i>MAX-HEAPIFY</i>?
<p>
<i>// Bubbling the ith element as high as possible.</i><br>
<cite><b>MIN-HEAPIFY(A, i)</b><br>
$\quad l\,=\,LEFT(i)$<br>
$\quad r\,=\,RIGHT(i)$<br>
$\quad if\,l\,\leq\,A.heap\_size\,and\,A[l]\,\lt\,A[i]$<br>
$\quad \quad smallest\,=\,l$<br>
$\quad else$<br>
$\quad \quad smallest\,=\,i$<br>
$\quad if\,r\,\leq\,A.heap\_size\,and\,A[r]\,\lt\,A[smallest]$<br>
$\quad \quad smallest\,=\,r$<br>
$\quad if\,smallest\,!=\,i$<br>
$\quad \quad SWAP\,A[i]\,and\,A[smallest]$<br>
$\quad \quad MIN-HEAPIFY(A,\,smallest)$<br>
</cite><br>
The running time is same as that of <i>MAX-HEAPIFY</i>, i.e., $O(\lg n)$.
</p>

### exe.6.2.3 What is the effect of calling <i>MAX-HEAPIFY(A, i)</i> when the element $A[i]$ is larger than its children?
<p>
No effect. Since the procedure will terminate directly, no swapping and recursing needed.
</p>

### exe.6.2.4 What is the effect of calling <i>MAX-HEAPIFY(A, i)</i> for $i\,\gt\,\frac{A.heap-size}{2}$?
<p>
No effect. In that case, both LEFT(i) and RIGHT(i) fail the comparison with A.heap-size and largest stores i, so that, swapping and recursing will not be performed.

### exe.6.2.6 Show that the worst-case running time of <i>MAX-HEAPIFY</i> on a heap of size n is $\Omega(\lg n)$. (Hint: For a heap with n nodes, give node values that cause <i>MAX-HEAPIFY</i> to be called recursively at every node on a simple path from the root down to a leaf.)
<p>Take the leftmost path in given heap, let the smallest element be the root and left-child is larger than right-child for every element, then <b>MAX-HEAPIFY</b> will be called for $h$ many times ($h$ is the height), since it is called at each level in order to sink the smallest element to the leftmost leaf. Since $h\,=\,\lfloor\lg n\rfloor$, the worst-case running time of the procedure is $\Omega(\lg n)$.
</p>
<p>
<u>Following solution is from "Algs, Instructor's Manual".</u><br>
<cite>
If you put a value at the root that is less than every value in the left and right subtrees, then <b>MAX-HEAPIFY</b> will be called recursively until a leaf is reached. To make the recursive calls travers the longest path to a leaf, choose values that make <b>MAX-HEAPIFY</b> always recurse on the left child. It follows the left branch when the left-child $\geq$ right-child, so putting 0 at the root and 1 at all the other nodes, for example, will accomplish taht. With such values, <b>MAX-HEAPIFY</b> will be called $h$ times (where $h$ is the height of heap, which is the number of edges in the longest path from the root to a leaf), so its running time will be $\Theta(h)$ (since each call does $\Theta(1)$ work), which is $\Theta(\lg n)$. Since we have a case in which <b>MAX-HEAPIFY</b>'s running time is $\Theta(\lg n)$, its worst-case running time is $\Omega(\lg n)$.
</cite>
</p>

### exe. 6.3.3 Show that there are at most $\lceil\frac{n}{2^{h\,+\,1}}\rceil$ nodes of height $h$ in any n-element heap.
<p>
<big><u>My solution (more mathematical)</u></big><br>
<big><b>Base Case:</b></big><br>
A max-heap can have at most $\frac{n\,+\,1}{2}$ nodes at the lowest level (when the heap is a complete binary tree), and<br>
<center>
$x\,=\,\frac{n\,+\,1}{2}\,\leq\,\lceil\frac{n}{2}\rceil\,=\lceil\frac{n}{2^{1}}\rceil\,=\,\lceil\frac{n}{2^{0\,+\,1}}\rceil\,=\,\lceil\frac{n}{2^{h\,+\,1}}\rceil$.
</center><br>
For non-complete binary trees, number of nodes is less than $\lceil\frac{n}{2}\rceil$. Thus, the base case holds.<br>
<big><b>Induction:</b></big><br>
Assume that it holds for nodes of height $(h\,-\,1)$. Then take a tree, remove all the nodes from the lowest level and get a new tree (which is a complete binary tree) $T^\prime$ with height $h^\prime\,=\,h\,-\,1$ and length $n^\prime\,=\,n\,-\,\lceil\frac{n}{2}\rceil\,=\,\lfloor\frac{n}{2}\rfloor$, and the number of nodes at the lowest level of $T^\prime$ is<br>
<center>
$\frac{n^\prime\,+\,1}{2}\,\leq\,\lceil\frac{n^\prime}{2^{h^\prime\,+\,1}}\rceil\,=\,\lceil\frac{\lfloor n/2 \rfloor}{2^{h\,-\,1\,+\,1}}\rceil\,\leq\,\lceil\frac{(n/2)}{2^{h}}\rceil\,=\,\lceil\frac{n}{2^{h\,+\,1}}\rceil$
</center><br>
Repeat above process through all level of the tree, and keep in mind that height decrease each round.
</p>
<p><big><b><u>KEY OF SOLUTION: height of node is the distance of the node to the lowest leaf!</u></b></big></p>
<p>
<cite>
<big><u>Following solution is from "Algs, Instructor's Manual". (more logical)</u></big><br>
<b>Proof by induction on $h$.</b><br>
<big><b>Basis:</b></big><br>
Show that it's TRUE for $h\,=\,0$ (i.e., that $\#\,of leaves\,\leq\,\lceil\frac{n}{2^{h+1}}\rceil\,=\,\lceil\frac{n}{2}\rceil$).<br>
The tree leaves (nodes at height 0) are at depths $H$ and $(H\,-\,1)$. They consist of <br>
$\diamondsuit$ all nodes at depth $H$, and<br>
$\diamondsuit$ the nodes at depth $(H\,+\,1)$ that are not parents of depth-$H$ nodes.<br>
Let $x$ be the number of nodes at depth $H$ -- that is, the number of nodes in the bottom (possibly incomplete) level.<br>
Note that $(n\,-\,x)$ is odd, because the $(n\,-\,x)$ nodes above the bottom level form a complete binary tree, and a complete binary tree has an odd number of nodes ($1$ less than a power of $2$). Thus if $n$ is odd, $x$ is even, otherwise, $x$ is odd.<br>
Now, proof the base case separately when $n$ is even and odd. <br>
<b>$n$ is ODD</b><br>
$\diamondsuit$ If $n$ is odd, then $x$ is even, so all nodes have siblings -- i.e., all internal nodes have 2 children. Thus $num\_internal\_nodes\,=\,num\_of\_leaves\,-\,1$. So, $n\,=\,num\_of\_leaves\,+\,num\_of\_internal\_nodes\,=\,2\,*\,num\_of\_leaves\,-\,1$. Thus, $num\_of\_leaves\,=\,\frac{n\,+\,1}{2}\,=\,\lceil\frac{n}{2}\rceil$ (The latter equality holds because n is odd).<br>
$\diamondsuit$ If $n$ is even, then $x$ is odd, and one leaf does not have sibling. If we gave it a sibling, then we would have $(n\,+\,1)$ nodes, where $x\prime\,=\,(n\,+\,1)$ is odd, so this case has been prooved. Observe that we would also increase the number of leaves by $1$, since we added a node to a parent that already had a child. By the odd-node case, $num\_of\_leaves\,+\,1\,=\,\lceil\frac{n\,+\,1}{2}\rceil\,=\,\lceil\frac{n}{2}\rceil\,+\,1$. (The latter equality holds since n is even.)<br>
Therefore, in either case, $num\_of\_leaves\,=\,\lceil\frac{n}{2}\rceil$.<br>
<big><b>Inductive step:</b></big><br>
Show that if it's TRUE for height $(h\,-\,1)$, it's TRUE for $h$.<br>
Let $n_h$ be the number of nodes at height $h$ in the n-node tree $T$.<br>
Consider the tree $T\prime$ formed by removing the leaves of $T$. It has $n\prime\,=\,n\,-\,n_0$ nodes. We know from the base case that $n_0\,=\,\lceil\frac{n}{2}\rceil\,$, so $n\prime\,=\,n\,-\,\lceil\frac{n}{2}\rceil\,=\,\lfloor\frac{n}{2}\rfloor$. Note that the nodes at height $h$ in $T$ would be at height $(h\,-\,1)$ if the leaves of the tree were removed -- that is, they are at height $(h\,-\,1)$ in $T\prime$. Letting $\sideset{}{_{h-1}^{\prime}}n$ denote the number of nodes at height $(h\,-\,1)$ in $T\prime$, we have $n_h\,=\,\sideset{}{_{h-1}^{\prime}}n$.<br>
By <big><b>inductoin</b></big>, we can bould $\sideset{}{_{h-1}^{\prime}}n$:<br>
<big>$n_h\,=\,\lceil\frac{\sideset{}{_{h-1}^{\prime}}n}{2}\rfloor\,\leq\,\lceil\frac{\lfloor\frac{n}{2}\rfloor}{2^h}\rceil\,\leq\,\lceil\frac{n/2}{2^h}\rceil\,=\,\lceil\frac{n}{2^{h+1}}\rceil$.</big>
</cite>
</p>
